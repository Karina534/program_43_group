1.3 Есть ли на ФИКТ однофамильцы? Однофамильцы – это люди, у которых полностью совпадают фамилии. Если да, выведите этих людей. Сколько их на каждом курсе?

```python
def find_one_surnames(df):
  uniq = set()
  for fio in data['фио']:
    surnmae = fio.split(' ')
    uniq.add(surnmae[0])

  for el in uniq:
    d = []
    for fio in data['фио']:
      surnmae = fio.split(' ')
      if surnmae[0] == el:
        d.append(fio)
    if len(d) > 1:
      print(d[:5])  # Вывод первых пяти однофамильцев для каждой фамилии
      print('\n')


df['фамилия'] = df['фио'].apply(lambda x: x.split(' ')[0])
single_last_names = df.groupby('фамилия')['фамилия'].size().reset_index(name='count')
single_last_names = single_last_names[single_last_names['count'] > 1]
result = df[df['фамилия'].isin(single_last_names['фамилия'])].groupby(['курс', 'фамилия'])['фамилия'].size().reset_index(name='count')
print(result)
```

2.1 Покажите на столбчатой диаграмме с группировкой, как отличается количество студентов на факультете инфокоммуникационных технологий (ФИКТ) и факультете информационных технологий и программирования (ФИТиП) на каждом курсе.

```python
import matplotlib.pyplot as plt
fict = df[df['факультет'] == 'факультет инфокоммуникационных технологий']
fitip = df[df['факультет'] == 'факультет информационных технологий и программирования']
fict_group = fict.groupby('курс').size().reset_index(name='count')
fitip_group = fict.groupby('курс').size().reset_index(name='count')

plt.bar(fitip_group['курс'], fitip_group['count'], label='ФИТиП')
plt.bar(fict_group['курс'], fict_group['count'], bottom=fitip_group['count'], label='ФИКТ')
plt.xlabel('Курс')
plt.ylabel('Количество студентов')
plt.title('Количество студентов на факультете')
plt.legend()
plt.show()
```

2.2 На круговой диаграмме покажите среднее количество студентов в группе каждого курса.

```python
r = df.groupby(['курс', 'группа'])['группа'].size().reset_index(name='count')
r2 = r.groupby('курс')['count'].median().reset_index(name='median')

fig = plt.figure(figsize=(10, 7))
plt.pie(r2['median'], labels=['1-й, Среднее количество учениеков в группе: ' + str(r2['median'][0]),
                              '2-й, Среднее количество учениеков в группе: ' + str(r2['median'][1]),
                              '3-й, Среднее количество учениеков в группе: ' + str(r2['median'][2]),
                              '4-й, Среднее количество учениеков в группе: ' + str(r2['median'][3])])
plt.title('Среднее количестов студентов в группе каждого курса')
plt.show()
```
